## Necessary Imports

```python
from fastai.vision.all import *
```

## Load Saved Model

```python
# Version
version = 2

# Images for Testing
test_images_path = "../test_data"

# Model Path
model_path = f"../models/pst_recognizer_v{version}.pkl"
```

```python
# Model Learn
model = load_learner(model_path)
```

## Verify Inference

### Inference - 1: Landscape Photography

```python
# Open the image
image = Image.open(f"{test_images_path}/landscape_photography.jpg")
image.to_thumb(128, 128)
```

```python
# Predict the type
pst = model.predict(f"{test_images_path}/landscape_photography.jpg")
print("Photographic Style Type: ", pst[0])
print("Index: ", pst[1])
print("Probability: ", pst[2][pst[1]])
print("All Probabilities: ", pst[2])
```
